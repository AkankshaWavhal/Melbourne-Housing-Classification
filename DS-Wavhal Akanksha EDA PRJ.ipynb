# KDE: Kernal Density Estimator

Kernal Density Estimation visualizes the distribution of data over a continuous interval.
The conventional scale for KDE is =  Total frequency of each bin * Probability.

```python
sns.histplot(data=df, x='Price',kde='True');
```
